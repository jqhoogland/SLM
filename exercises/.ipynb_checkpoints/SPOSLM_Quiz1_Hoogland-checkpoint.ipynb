```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
SIGMA = 10
B = 8/3
RHO = 40
X_0 = np.array([10., 10., 10.])
```

```python
def runge_kutta_method(x_0, f, delta_t, n_steps):
    xs = np.zeros((n_steps, 3))
    xs[0] = x_0

    for step in range(1, n_steps):
        k_1 = f(xs[step-1]) * delta_t
        k_2 = f(xs[step-1] + k_1 / 2) * delta_t
        k_3 = f(xs[step-1] + k_2 / 2) * delta_t
        k_4 = f(xs[step-1] + k_3) * delta_t
        xs[step] = xs[step-1] + ((k_1 + 2 * k_2 + 2 * k_3 + k_4) / 6)

    return xs
```

```python
def euler_method(x_0, f, delta_t, n_steps):
    xs = np.zeros((n_steps, 3))
    xs[0] = x_0

    for step in range(1, n_steps):
        xs[step] = xs[step-1]  + f(xs[step-1]) * delta_t

    return xs
```

```python
def lorenz_equations(x, sigma, rho, b):
    return np.array([
        sigma * (x[1] - x[0]),
        rho * x[0] - x[1] - x[0] * x[2],
        x[0] * x[1] - b * x[2]
    ])
```

```python
def lorenz_trajectory(x_0, delta_t, n_steps, sigma=SIGMA, rho=RHO, b=B):
    _lorenz_equations = lambda x: lorenz_equations(x, sigma, rho, b)
    return runge_kutta_method(x_0, _lorenz_equations, delta_t, n_steps)
```

```python
print('rho={}, sigma={}, b={}'.format(RHO,SIGMA,B))
lorenz_sequence = lorenz_trajectory(X_0, 0.01, 1000)

plt.plot(lorenz_sequence[:, 0], lorenz_sequence[:, 2])
```

```python
rho_1 = 0.5

print('rho={}, sigma={}, b={}'.format(rho_1 ,SIGMA,B))
lorenz_sequence = lorenz_trajectory(X_0, 0.01, 1000, rho_1)

plt.plot(lorenz_sequence[:, 0], lorenz_sequence[:, 2])
```

```python
rho_2 = 1.5

print('rho={}, sigma={}, b={}'.format(rho_2 ,SIGMA,B))
lorenz_sequence = lorenz_trajectory(X_0, 0.01, 1000, rho_2)

plt.plot(lorenz_sequence[:, 0], lorenz_sequence[:, 2])
```

```python
rho_3 = 1.5

print('Different initial conditions -> different fixed point: rho={}, sigma={}, b={}'.format(rho_3 ,SIGMA,B))
lorenz_sequence = lorenz_trajectory(np.array([-10.,-10.,-10.]), 0.01, 1000, rho_3)

plt.plot(lorenz_sequence[:, 0], lorenz_sequence[:, 2])
```

```python
rho_4 = 28

print('Lorenz\'s original paramter choice: rho={}, sigma={}, b={}'.format(rho_4,SIGMA,B))
lorenz_sequence = lorenz_trajectory(X_0, 0.01, 1000, rho_4)

plt.plot(lorenz_sequence[:, 0], lorenz_sequence[:, 2])
```
